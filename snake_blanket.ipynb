```python
def get_color(n,alpha): #
    beta=1/120
    if n<121:
        b=0
        r=120-n
        g=n
    if (n>120)&(n<241):
        r=0
        b=n-120
        g=240-n
    if n>240:
        r=n-240
        g=0
        b=360-n
    return (r*beta,b*beta,g*beta,alpha) #
```
